```python
import pandas as pd
import matplotlib.pyplot as plot
import os, sys
import datetime
```

```python
merged_df = pd.read_csv('/content/merged_df.csv')
```

```python
first_query = merged_df[["created_at_hospital", "patient_id"]].groupby('created_at_hospital').count().add_prefix("No_of_").reset_index()
```

```python
second_query = merged_df[["patient_id", "patient_name", 'doctor_id', 'hospital_visit_month']].groupby(['doctor_id', 'hospital_visit_month',"patient_name"]).count().add_prefix("No_of_patient_id")
```

```python
second_query
```

```python
res = merged_df[["patient_id", 'sex', 'hospital_visit_month']].groupby(['hospital_visit_month',"sex"]).count().add_prefix("No_of_").reset_index()
		
res = res.pivot(index='hospital_visit_month', columns='sex', values='No_of_patient_id')
		
res = (res['female'] / res['male']).rename('ratio').reset_index()
```
